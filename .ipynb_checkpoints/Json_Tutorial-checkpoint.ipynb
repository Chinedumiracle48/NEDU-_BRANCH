```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = {'English': [], 'Maths': [], 'Physics': [], 'Chemistry': [], 'Biology': []}

    def add_grade(self, subject, grade):
        if subject in self.grades:
            self.grades[subject].append(grade)
        else:
            print(f"Subject '{subject}' not found. Please choose from: {', '.join(self.grades.keys())}")

    def calculate_average_grade(self, subject):
        if subject in self.grades and len(self.grades[subject]) > 0:
            return sum(self.grades[subject]) / len(self.grades[subject])
        else:
            return None

    def __str__(self):
        return f"Student Name: {self.name}\nStudent ID: {self.student_id}\nGrades: {self.grades}"


# Example usage:
student1 = Student("John Smith", "S12345")
student1.add_grade("English", 85)
student1.add_grade("Maths", 90)
student1.add_grade("Physics", 78)
student1.add_grade("Chemistry", 92)
student1.add_grade("Biology", 88)

print(student1)
print(f"Average English Grade: {student1.calculate_average_grade('English')}")
print(f"Average Maths Grade: {student1.calculate_average_grade('Maths')}")
```

```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
plt.plot(input_values, squares, linewidth=5)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Values", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(2, 4, s=150)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Values", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values, y_values, edgecolor="none", s=40)
# plt.axis([0, 1100, 0, 1100000])
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = list(range(1001))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Greens,
    edgecolor='none', s=40)
plt.show()
```

```python
# 15-1. Cubes: A number raised to the third power is a cube. Plot the first five
# cubic numbers, and then plot the first 5000 cubic numbers.
# # 15-2. Colored Cubes: Apply a colormap to your cubes plot.
import matplotlib.pyplot as plt

x_values = list(range(5000))
y_values = [x**3 for x in x_values]
plt.scatter(x_values, y_values, edgecolor="none", c=y_values, cmap=plt.cm.Reds, s = 40)
plt.show()
```
